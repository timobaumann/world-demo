```python
import numpy as np

import matplotlib.pyplot as plt

import soundfile as sf
import pyworld as pw

import IPython
import matplotlib      # Remove this line if you don't need them
matplotlib.use('Agg')  # Remove this line if you don't need them
```

```python
#parser = argparse.ArgumentParser()
#parser.add_argument("-f", "--frame_period", type=float, default=5.0)
#parser.add_argument("-s", "--speed", type=int, default=1)
FRAME_PERIOD = 5.0
SPEED=1
EPSILON = 1e-8
```

```python
x, fs = sf.read('utterance/test234.wav')
# x : the actual samples
# fs : sampling rate of the audio
IPython.display.Audio(data=x, rate=fs)
```

```python
f0, sp, ap = pw.wav2world(x, fs)    # use default options
```

```python
y = pw.synthesize(f0, sp, ap, fs, pw.default_frame_period)
IPython.display.Audio(data=y, rate=fs)
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import warnings
warnings.filterwarnings('ignore')
def syn(f0_shift): # f0_manip should be given in CENT (100th of a semi-tone)
    CENT_CONST = 1731.2340490667560888319096172 # 1200 / ln(2)
    new_f0 = np.exp((np.log(f0/110)*CENT_CONST + f0_shift) / CENT_CONST) * 110
    y = pw.synthesize(new_f0, sp, ap, fs, pw.default_frame_period)
    return IPython.display.Audio(data=y, rate=fs)
interact(syn, f0_shift=widgets.IntSlider(min=-1200, max=1200, step=1, value=0));
```

Mit dem Widget oben kann man f<sub>0</sub> nach oben/unten verschieben. Die Einheit ist Cent (also Hundertstel Halbtne).
